```python
def add(a,b):
    return a+b

def myreduce(add,seq):
    a = seq[0]
    for i in range(1,len(seq)):
        b = seq[i]
        a = add(a,b)
    
    return a
        
l = [10,15,25,40,10]

myreduce(add,l)
```
